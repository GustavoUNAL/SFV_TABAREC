## Etapa 4 — Dimensionamiento

**Estado:** ⬜ Pendiente

Cálculo de potencia pico y cantidad de módulos.

```python
# --- Parámetros de dimensionamiento ---
PR = 0.80          # Performance Ratio (típico 0.75–0.85)
HSP = None         # Horas Sol Pico diarias (kWh/m²/día)
fp = None          # Factor de pérdidas adicionales (opcional)

# Módulo seleccionado (se completará en Etapa 5)
modulo = {
    "marca": "",
    "modelo": "",
    "potencia_wp": None,       # Wp por módulo
    "voc": None,               # Voltaje circuito abierto (V)
    "isc": None,               # Corriente cortocircuito (A)
    "vmp": None,               # Voltaje punto máxima potencia (V)
    "imp": None,               # Corriente punto máxima potencia (A)
    "eficiencia_pct": None,
    "dimensiones_mm": (None, None, None),  # L x W x H
}

def dimensionar_sfv(consumo_anual_kwh, HSP, potencia_wp, PR=0.80, cobertura=1.0):
    """
    Dimensiona un SFV según consumo anual.
    
    Ppico (kWp) = Consumo_anual / (HSP × 365 × PR) × cobertura
    N_módulos = Ppico × 1000 / potencia_wp
    """
    if not all([consumo_anual_kwh, HSP, potencia_wp]):
        return None
    
    ppico_kw = (consumo_anual_kwh / (HSP * 365 * PR)) * cobertura
    n_modulos = ppico_kw * 1000 / potencia_wp
    n_modulos_redondeado = round(n_modulos)
    ppico_real_kw = n_modulos_redondeado * potencia_wp / 1000
    
    return {
        "potencia_pico_kw": round(ppico_kw, 2),
        "n_modulos_calculado": round(n_modulos, 1),
        "n_modulos": n_modulos_redondeado,
        "potencia_real_kw": round(ppico_real_kw, 2),
    }

# Ejecutar cuando tengamos los datos
resultado_dimensionamiento = dimensionar_sfv(
    consumo.get("consumo_anual_kwh"),
    HSP,
    modulo.get("potencia_wp"),
    PR=PR,
    cobertura=(objetivo.get("cobertura_objetivo_pct") or 100) / 100,
)

if resultado_dimensionamiento:
    for k, v in resultado_dimensionamiento.items():
        print(f"{k}: {v}")
else:
    print("Dimensionamiento: completar consumo, HSP y potencia del módulo")
```

## Etapa 6 — Diseño eléctrico

**Estado:** ⬜ Pendiente

Configuración de strings, verificación de rangos MPPT y caída de tensión.

```python
# --- Diseño eléctrico ---
diseno_electrico = {
    "n_strings": None,
    "modulos_por_string": None,
    "voc_string_frio": None,       # Voc × N × factor temperatura frío
    "vmp_string_caliente": None,   # Vmp × N × factor temperatura calor
    "isc_string": None,
    "cable_dc": {
        "seccion_mm2": None,
        "longitud_m": None,
        "caida_tension_pct": None,  # Objetivo < 3%
    },
    "cable_ac": {
        "seccion_mm2": None,
        "longitud_m": None,
        "caida_tension_pct": None,
    },
}

def verificar_string(n_modulos_string, voc_mod, vmp_mod, voc_max_inversor, vmp_min_inversor,
                       temp_min=0, temp_max=45, coef_voc=-0.003, coef_vmp=-0.004):
    """Verifica que el string esté dentro del rango MPPT del inversor."""
    voc_frio = n_modulos_string * voc_mod * (1 + coef_voc * (temp_min - 25))
    vmp_caliente = n_modulos_string * vmp_mod * (1 + coef_vmp * (temp_max - 25))
    
    ok_voc = voc_frio <= voc_max_inversor
    ok_vmp = vmp_caliente >= vmp_min_inversor
    
    return {
        "voc_frio_v": round(voc_frio, 1),
        "vmp_caliente_v": round(vmp_caliente, 1),
        "voc_ok": ok_voc,
        "vmp_ok": ok_vmp,
        "string_valido": ok_voc and ok_vmp,
    }

print("Diseño eléctrico: por calcular")
```

## Etapa 8 — Estimación económica

**Estado:** ⬜ Pendiente

```python
# --- Estimación económica ---
economia = {
    "costo_modulos": None,
    "costo_inversor": None,
    "costo_estructura": None,
    "costo_instalacion": None,
    "costo_protecciones_cableado": None,
    "costo_baterias": None,
    "costo_tramites": None,
    "moneda": "USD",               # o ARS, etc.
}

def calcular_capex(economia):
    campos = ["costo_modulos", "costo_inversor", "costo_estructura",
              "costo_instalacion", "costo_protecciones_cableado",
              "costo_baterias", "costo_tramites"]
    total = sum(v for k, v in economia.items() if k in campos and v)
    return total if total > 0 else None

def calcular_payback(capex, ahorro_anual):
    if capex and ahorro_anual and ahorro_anual > 0:
        return round(capex / ahorro_anual, 1)
    return None

capex = calcular_capex(economia)
print(f"CAPEX total: {capex or 'por calcular'} {economia['moneda']}")
```

## Etapa 9 — Producción estimada

**Estado:** ⬜ Pendiente

```python
# --- Producción estimada ---
def estimar_produccion(ppico_kw, HSP, PR=0.80):
    """Estima producción anual en kWh."""
    if not all([ppico_kw, HSP]):
        return None
    produccion_anual = ppico_kw * HSP * 365 * PR
    factor_planta = (produccion_anual / (ppico_kw * 8760)) * 100
    return {
        "produccion_anual_kwh": round(produccion_anual, 0),
        "produccion_mensual_prom_kwh": round(produccion_anual / 12, 0),
        "factor_planta_pct": round(factor_planta, 1),
    }

ppico = resultado_dimensionamiento["potencia_real_kw"] if resultado_dimensionamiento else None
produccion = estimar_produccion(ppico, HSP, PR)

if produccion:
    for k, v in produccion.items():
        print(f"{k}: {v}")
else:
    print("Producción: completar potencia pico y HSP")
```
